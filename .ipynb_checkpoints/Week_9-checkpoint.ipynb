## 03 Solution 01

```python
import pandas as pd
data = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'

df = pd.read_csv(data)

h = df.head()
h
```

## 04 Solution 02

```python
honda = df[df['Identification.Make'] == 'Honda']

gas = df[df['Fuel Information.Fuel Type'] == "Gasoline"]
            
outliers = df[df['Engine Information.Engine Statistics.Horsepower'] < df['Engine Information.Engine Statistics.Horsepower'].quantile(0.90)]

double_feature = df[['Fuel Information.Highway mpg', 'Engine Information.Engine Statistics.Horsepower']]

from sklearn import preprocessing
scalar = preprocessing.MinMaxScaler()
names = double_feature.columns
data = scalar.fit_transform(double_feature)
normalized_df = pd.DataFrame(data, columns=names)
normalized_df
```

## 05 Solution 03

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

x = normalized_df['Engine Information.Engine Statistics.Horsepower']
x_reshape = np.array(x).reshape(-1,1)

y = normalized_df['Fuel Information.Highway mpg']

model = LinearRegression()
model.fit(x_reshape, y)

coef = model.coef_
print("coefficient: " + str(coef))

pred_y = model.predict([[1]])
print("y predicted by x = 1: " + str(pred_y))

normalized_df.plot.scatter(x=0, y=1)
m, b = np.polyfit(x,y,1)

plt.plot(x,coef*x+b)
```
